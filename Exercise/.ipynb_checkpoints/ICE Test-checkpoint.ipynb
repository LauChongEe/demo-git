```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('../../data/week04/weather.csv')
data
```

```python
#convert date column to real date format
data = data[data['s_id']=='USW00094889']
data["date"] = pd.to_datetime(data["date"])

data.head()
```

```python
#sort the value by station identification code and date
data.sort_values(['s_id','date'], ascending=True, inplace=True)

#convert date string to date
convert = lambda x:x*0.1
data['temp'] = data['value'].apply(convert)
data
```

```python
# remove the leap year date, 29th February & 2 dates
#data1 = data1[(data1['date'] != pd.to_datetime('01/04/2006')) & (data1['date'] != pd.to_datetime('01/05/2006'))]

data1 = data[(data['date'].dt.year != 2015) & ~((data['date'].dt.day == 29) & (data['date'].dt.month == 2))]
data1
```

```python
#spread out month, day and year
# date_index=pd.DatetimeIndex(data1['date'])
# data1['month']=date_index.month
# data1['day']=date_index.day
# data1['year']=date_index.year
# data1['dayofyear']=date_index.dayofyear

#create df for TMAX Elements and get max value
df_tmax = data1[data1['element_type']=='TMAX']
df_tmax.set_index('date', inplace=True) # change index to date before perform resample
df_tmax = df_tmax.resample('D').asfreq() #fill blank to NaN
#tmax = pd.pivot_table(df_tmax,index='dayofyear',values=['temp'],aggfunc={'temp':'max'}) #get max value per day of each year
tmax = df_tmax.groupby(df_tmax.index.dayofyear)[['temp']].max() #get max value per day of each year, groupby dayofyear mean it will become index

#create df for TMIN Elements and get min value
df_tmin = data1[data1['element_type']=='TMIN']
df_tmin.set_index('date', inplace=True) # change index to date before perform resample
df_tmin = df_tmin.resample('D').asfreq() #fill blank to NaN
#tmin=pd.pivot_table(df_tmin,index='dayofyear',values=['temp'],aggfunc={'temp':'min'}) #get min value per day of each year
tmin = df_tmin.groupby(df_tmin.index.dayofyear)[['temp']].min() #get max value per day of each year, groupby dayofyear mean it will become index
### dayofyear will always assign a number to 29/Feb

tmax.shape
tmin.shape

#data1[(data1['date'].dt.year == 2008) & (data1['element_type']=='TMAX')].iloc[31:60]

```

```python
# plot a line graph of the record high and record low 
# temperatures by day of the year over the period 2005-2014

#create figure and axes
fig = plt.figure()
fig, ax = plt.subplots(figsize=(20,10))
plt.plot(tmin.temp,color="red") #plt.plot(tmin.index, tmin.temp,color="red")
plt.plot(tmax.temp,color="blue") #plt.plot(tmax.index, tmax.temp,color="blue")
plt.fill_between(tmax.index, tmax.temp, tmin.temp, color='yellow', alpha=0.5)

plt.title('record high and record low temperatures by day of the year over the period 2005-2014')
plt.xlabel('Year')
plt.ylabel('Temperature')
```

```python
# get 2015 data only and exclude 29/Feb

data2 = data[(data['date'].dt.year == 2015) & ~((data['date'].dt.day == 29) & (data['date'].dt.month == 2))]
data2
```

```python
#create df for TMAX Elements and get max value
df_tmax_2 = data2[data2['element_type']=='TMAX']
df_tmax_2.set_index('date', inplace=True) # change index to date before perform resample
df_tmax_2 = df_tmax_2.resample('D').asfreq() #fill blank to NaN
#tmax = pd.pivot_table(df_tmax,index='dayofyear',values=['temp'],aggfunc={'temp':'max'}) #get max value per day of each year
tmax_2 = df_tmax_2.groupby(df_tmax_2.index.dayofyear)[['temp']].max() #get max value per day of each year, groupby dayofyear mean it will become index
tmax_2.loc[366] = [None]

#create df for TMIN Elements and get min value
df_tmin_2 = data2[data2['element_type']=='TMIN']
df_tmin_2.set_index('date', inplace=True) # change index to date before perform resample
df_tmin_2 = df_tmin_2.resample('D').asfreq() #fill blank to NaN
#tmin=pd.pivot_table(df_tmin,index='dayofyear',values=['temp'],aggfunc={'temp':'min'}) #get min value per day of each year
tmin_2 = df_tmin_2.groupby(df_tmin_2.index.dayofyear)[['temp']].min() #get max value per day of each year, groupby dayofyear mean it will become index
tmin_2.loc[366] = [None]


result_max = pd.concat([tmax.rename(columns={'temp':'2005-2014'}), tmax_2.rename(columns={'temp':'2015'})], axis=1)
result_max[result_max['2015'] > result_max['2005-2014']]

result_min = pd.concat([tmin.rename(columns={'temp':'2005-2014'}), tmin_2.rename(columns={'temp':'2015'})], axis=1)
result_min[result_min['2015'] < result_min['2005-2014']]
```

```python
# find record breaking high & low temperature in year 2015 compared to year 2005-2014
tmax_2 = tmax_2[(tmax_2['temp'] > tmax['temp'])]
tmax_2
tmin_2 = tmin_2[(tmin_2['temp'] < tmin['temp'])]
tmin_2
```

```python
# Plot scatter plot of the 2015 data for any points (highs and lows)
# for which the ten year record (2005-2014) record high or record low 
# was broken in 2015

#create figure and axes
fig = plt.figure()
fig, ax = plt.subplots(figsize=(20,20))
plt.scatter(tmin_2.index, tmin_2.temp,color="green")
plt.scatter(tmax_2.index, tmax_2.temp,color="purple") 

plt.plot(tmin.temp,color="red") #plt.plot(tmin.index, tmin.temp,color="red")
plt.plot(tmax.temp,color="blue") #plt.plot(tmax.index, tmax.temp,color="blue")
plt.fill_between(tmax.index, tmax.temp, tmin.temp, color='yellow', alpha=0.5)

plt.title('2015 data for any points (highs and lows) broke record from year 2005-2014')
plt.xlabel('Year')
plt.ylabel('Temperature')
```
